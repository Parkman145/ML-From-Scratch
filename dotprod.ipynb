```python
import numpy as np
```

```python
input_samples = 5
sample_size = 3
inp = np.arange(input_samples*sample_size).reshape((input_samples, sample_size))+1
# inp = np.arange(sample_size)+1

weights = np.array([
    [0.5, 0.25],
    [1, 1],
    [2, 2]])
print(weights.shape)
print("Input")
print(inp)
print("Weights")
print(weights)
print("Result")
print(np.dot(inp, weights))
```
