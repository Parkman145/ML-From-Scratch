```python
input_samples = 5
sample_size = 3
inp = np.arange(input_samples*sample_size).reshape((input_samples, sample_size))+1
# inp = np.arange(sample_size)+1

weights = np.array([
    [0.5, 0.25],
    [1, 1],
    [2, 2]])
print(weights.shape)
print("Input")
print(inp)
print("Weights")
print(weights)
print("Result")
print(np.dot(inp, weights))
```

```python
n = 2
m = 3
p = 4
a = np.arange(n*m).reshape((m, n))
b = np.arange(p*n).reshape((n, p))
print(a, end="\n\n")
print(b, end="\n\n")
print(np.matmul(a,b))
```

```python
def matmul(a, b):
    m = a.shape[0]
    p = b.shape[1]
    output = np.zeros(shape=(m, p))

    for i in range(m):
        for j in range(p):
            output[i, j] = np.sum(a[i]*b[:,j])

    return output

n = 2
m = 3
p = 3
a = np.arange(n*m).reshape((m, n))
b = np.arange(p*n).reshape((n, p))
print(a, end="\n\n")
print(b, end="\n\n")
print(matmul(a,b))
```
